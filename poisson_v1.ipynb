```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math 
```

```python
s = np.random.poisson(3, (1000,1000))
```

```python
s=pd.DataFrame(s)
```

```python
s.shape
```

```python
s.head()
```

```python
col_average=[]
for i in range(s.shape[1]):
    col_average.append(s.iloc[:,i].mean())
```

```python
col_average
```

```python
col_std=[]
for i in range(s.shape[1]):
    col_std.append(s.iloc[:,i].std())
```

```python
col_std
```

```python
average_col_mean=np.mean(col_average)
```

```python
average_col_mean
```

```python
average_std_mean=np.mean(col_std)
```

```python
average_std_mean
```

```python
sample_std_dev=average_std_mean/math.sqrt(1000)
```

```python
sample_std_dev
```
