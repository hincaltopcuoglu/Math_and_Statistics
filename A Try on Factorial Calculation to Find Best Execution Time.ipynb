```python
import numpy as np
from functools import reduce
import time
import math
import timeit
import operator
import cProfile
```

### lets make Factorial Calculation by defining our own function and measure the executing time between other approaches

### Firstly, let's start with Python's own implementation which is in Math library and defined as factorial(). This library and function is written in C Language

### Make a big Calculation

```python
math.factorial(1000)
```

### let's look at an approach which I see in Stackoverflow. It uses while condition approach

##### _sof = StackOverFlow

```python
def factorial_sof(n):
    num = 1
    while n >= 1:
        num = num * n
        n = n - 1
    return num
```

```python
factorial_sof(1000)
```

### Yes, it gives the same result as in math.factorial function

### Now lets define our function in old way, using for loop and if else condition and list

```python
def factorial_Hincal1(x) -> int():
    lst = []
    for i in range(x+1):
        if i == 0:
            lst.append(1)
        elif i == 1:
            lst.append(1)
        else:
            lst.append(i)
    return reduce(lambda x, y: x*y, lst)
```

```python
factorial_Hincal1(1000)
```

### Yes ! The Result is true with others !

### now lets check their execution times to compare which has most power

```python
cProfile.run('math.factorial(1000)')
```

#### WOOW , Python's own library have an execution time 0 seconds with only 4 function calls !

### Now Lets look at the function in StackOverFlow

```python
cProfile.run('factorial_sof(1000)')
```

#### _sof function with while loop also do very well job with only millisecond difference and 4 function calls !

### Now check our function

```python
cProfile.run('factorial_Hincal1(1000)')
```

#### 2006 function calls.. So we must optimize it, lets try different approaches..

### NoW Lets make our function shortened in otherway and check results

```python
def factorial_Hincal2(x) -> int():
    lst = []
    for i in range(1,x+1):
        lst.append(i)
    return reduce(lambda x, y: x*y, lst)
```

### First Check if the result is true

```python
factorial_Hincal2(1000)
```

### Now Calculate our last function's execution time

```python
cProfile.run('factorial_Hincal2(1000)')
```

#### it only downgraded function calls 2 times..So lets continue to find best optimization

```python
def factorial_Hincal3(x) -> int():
    
    return  reduce(lambda a, b: a*b,range(1, x+1))    
```

```python
factorial_Hincal3(1000)
```

```python
cProfile.run('factorial_Hincal3(1000)')
```

#### it function calls decreases to the half of previous one, now the stats above show us that there is an optimization need for lambda function..

#### let's continue to try to find best optimization..here I will try mul operator approach

```python
def factorial_Hincal4(x) -> int():
    
    return reduce(operator.mul,range(1, x+1))      
```

```python
factorial_Hincal4(1000)
```

```python
cProfile.run('factorial_Hincal4(1000)')
```

#### WOOW , the mul operator with shortened range syntax reduced function calls to only 5 !

#### Lets continue to try. Our aim is to catch the result which is the best one : python's math.factorial() function which is written in C Language

```python
def factorial_Hincal5(x) -> int():
    num = 1
    for i in range(1,x+1):
        num *= i
    return num
```

```python
factorial_Hincal5(1000)
```

```python
cProfile.run('factorial_Hincal5(1000)')
```
